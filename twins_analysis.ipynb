```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_excel('Dados_Pesquisa_Presencial_Painel_USP.xlsx')
```

```python
# Contar a frequência de cada cronotipo
peak_counts = dados['Peak'].value_counts()

# Plotar o gráfico
peak_counts.plot(kind='bar')
plt.xlabel('Picos')
plt.ylabel('Count')
plt.title('Distribuição dos Picos')
plt.show()
```

```python
# Contar a frequência de cada cronotipo
chronotype_counts = dados['Chronotype'].value_counts()

# Plotar o gráfico
chronotype_counts.plot(kind='bar')
plt.xlabel('Chronotype')
plt.ylabel('Count')
plt.title('Frequência de cada Cronotipo')
plt.show()
```

```python
# Contar a quantidade de MZ e DZ para cada cronotipo
counts = dados.groupby(['Chronotype', 'DNA']).size().unstack(fill_value=0)

# Plotar o gráfico
counts.plot(kind='bar', stacked=True)
plt.xlabel('Cronotipo')
plt.ylabel('Count')
plt.title('Distribuição MZ e DZ no Cronotipo')
plt.show()
```

```python
# Contar a quantidade de MZ e DZ para cada cronotipo
counts = dados.groupby(['Sleep Quality', 'DNA']).size().unstack(fill_value=0)

# Plotar o gráfico
counts.plot(kind='bar', stacked=True)
plt.xlabel('Qualidade do Sono')
plt.ylabel('Count')
plt.title('Distribuição MZ e DZ na Qualidade do Sono')
plt.show()
```

```python
# Contar a quantidade de MZ e DZ para cada cronotipo
counts = dados.groupby(['Sleep length', 'DNA']).size().unstack(fill_value=0)

# Plotar o gráfico
counts.plot(kind='bar', stacked=True)
plt.xlabel('Duração do Sono')
plt.ylabel('Count')
plt.title('Quantidade de gêmeos MZ e DZ na Duração do Sono')
plt.show()
```

```python
# Contar a quantidade de MZ e DZ para cada cronotipo
counts = dados.groupby(['Sleep dysfunction', 'DNA']).size().unstack(fill_value=0)

# Plotar o gráfico
counts.plot(kind='bar', stacked=True)
plt.xlabel('Disfunção do Sono')
plt.ylabel('Count')
plt.title('Quantidade de gêmeos MZ e DZ em Disfunções do Sono')
plt.show()
```

```python
# Gráfico de barras para Chronotype e Sleep Quality
sns.countplot(x='Chronotype', hue='Sleep Quality', data=dados)
plt.xlabel('Chronotype')
plt.ylabel('Count')
plt.title('Distribuição de Sleep Quality por Chronotype')
plt.show()
```
